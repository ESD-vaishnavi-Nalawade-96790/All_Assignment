```python
import numpy as np
```

```python
def function1():
    l1 = [10,20,30,40,50]
    print(l1)
    print(f"Type = {type(l1)}")
    print(f"l1[0] = {type(l1[0])}")

function1()
```

```python
def function2():
    n_arr =np.array([11,22,33,44,55])
    print(n_arr)
    print(f"Type = {type(n_arr)}")
function2()
```

```python
def function3():
    n_arr = np.array([10,20,30,40,50,60])
    print(n_arr)
    print(f"Type = {type(n_arr)}")
    print(f"DType = {n_arr.dtype}")
    print(f"Item size = {n_arr.itemsize}")
    print(f"size = {n_arr.size}")
    print(f"nbytes = {n_arr.nbytes}")
    print(f"dim = {n_arr.ndim}")
function3()    
```
